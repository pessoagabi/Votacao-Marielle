```python
import csv
import pandas as pd
import numpy as np
```

```python
tabela = pd.read_csv("votacao-rj.csv", dtype={'zona':np.str}, sep =";", encoding="latin-1")
```

```python
tabela.head()
```

```python
#Começo a agrupar os dados e filtrar pela zona 1, que fica perto da zona que inclui a Maré:
zona1 = tabela.loc[tabela['zona'] == '1']
zona1.head()
```

```python
zona1_agg = zona1.groupby(['candidato','partido','situacao']).agg({'votos': sum}).sort_values(by='votos', ascending = False).reset_index()
zona1_agg.head(20)
```
